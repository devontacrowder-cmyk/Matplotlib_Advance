```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Expense_Categories =['Groceries', 'Utilities', 'Transportation', 'Dining Out', 'Entertainment'] 
Amount_Spent = [500, 300, 200, 400, 250]

plt.figure(figsize=(9, 6))
plt.grid(axis='y', linestyle='--', alpha=0.7, zorder=0)
plt.bar(Expense_Categories, Amount_Spent, width=0.2, color=['blue', 'orange', 'green', 'red', 'purple'], zorder=3)
plt.xlabel('Expense Categories')
plt.ylabel('Amount Spent ($)')
plt.title('Monthly Expense Breakdown')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
