# Bar Charts

-`plt.bar()` -

-`plt.barh()` - Horizontal bar display

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
CoursesName = ['C Course', 'C++ Course', 'Java Course', 'Python Course']
Student = [90, 60, 130, 190]


plt.figure(figsize=(9, 6))
plt.grid(True, axis='y', zorder=0)
plt.bar(CoursesName, Student, color='Red', width=0.5, edgecolor='Blue', zorder=3)
plt.xlabel('Courses Name')
plt.ylabel('Number of Students')
plt.title('Number of Students per Course')
plt.show()
```

```python
plt.barh(CoursesName, Student, color='Black')
plt.ylabel('Courses Name')
plt.xlabel('Number of Students')
plt.title('Number of Students per Course')
plt.show()
```

```python
male = [90, 60, 130, 600]
female = [100, 30, 160, 690]
plt.figure(figsize=(9, 6))
plt.grid(True, axis='y', zorder=0)
plt.bar(CoursesName, male, width=0.5, label='Male', zorder=3)
plt.bar(CoursesName, female, bottom=male, width=0.5, label='Female', zorder=3)
plt.xlabel('Courses Name')
plt.ylabel('Number of Students')
plt.title('Number of Students per Course')
plt.legend()
plt.show()


```

```python
years = ['2016', '2017', '2018', '2019']
rice_price = [57, 50, 47, 30]
wheat_price = [43, 50, 53, 70]


plt.figure(figsize=(9, 6))
plt.grid(True, axis='y', zorder=0)
plt.bar(years, rice_price, width=0.5, label='Rice Price', zorder=3, color='Yellow')
plt.bar(years, wheat_price, bottom=rice_price, width=0.5, label='Wheat Price', zorder=3, color='Pink')
plt.xlabel('Years')
plt.ylabel('Price')
plt.title('Price of Rice and Wheat over Time')
plt.legend()
plt.show()
```

```python
mustard =  np.array([10, 16, 13, 15])
rice_price = [57, 50, 47, 30]
wheat_price = [43, 50, 53, 70]

plt.figure(figsize=(9, 6))
plt.grid(True, axis='y', zorder=0)
plt.bar(years, rice_price, width=0.5, label='Rice', zorder=3, color='Yellow')
plt.bar(years, wheat_price, bottom=rice_price, width=0.5, label='Wheat', zorder=3, color='Pink')
plt.bar(years, mustard, label='Mustard', zorder=3, color='Green', width=0.5)
plt.xlabel('Years')
plt.ylabel('Grains Frequency')
plt.title('Price of Rice, Wheat and Mustard by Year')
plt.legend()
plt.show()
```

```python

# Convert lists to arrays so stacking works
rice_price = np.array(rice_price)
wheat_price = np.array(wheat_price)
mustard = np.array(mustard)
years = np.array(years)

plt.figure(figsize=(9, 6))
plt.grid(True, axis='y', zorder=0)

plt.bar(years, rice_price, width=0.5, label='Rice', zorder=3, color='yellow')

plt.bar(years, wheat_price, bottom=rice_price,
        width=0.5, label='Wheat', zorder=3, color='pink')

plt.bar(years, mustard, bottom=rice_price + wheat_price,
        width=0.5, label='Mustard', zorder=3, color='green')

plt.xlabel('Years')
plt.ylabel('Price')
plt.title('Price of Rice, Wheat and Mustard over Time')
plt.legend()
plt.show()
```

```python
CoursesName = ['C Course', 'C++ Course', 'Java Course', 'Python Course']
StudentGroup1 = [90, 60, 130, 600]
StudentGroup2 = [100, 30, 160, 690]

bar_width = 0.35
x = np.arange(len(CoursesName))



plt.figure(figsize=(9, 6))
plt.grid(True, axis='y', zorder=0)
plt.bar(x - bar_width/2, StudentGroup1, width=bar_width, label='Student Group 1', zorder=3, color='Black')
plt.bar(x + bar_width/2, StudentGroup2, width=bar_width, label='Student Group 2', zorder=3, color='Grey')
plt.xticks(x, CoursesName, rotation=45)
plt.title('Number of Students per Course by Group')
plt.legend()
plt.show()
```

```python
data = {'CoursesName': ['C Programming Course', 'C++ Programming Course', 'Java Programming Course', 'Python Programming Course'],
        'Student Group 1': [90, 60, 130, 600],
        'Student Group 2': [100, 30, 160, 690]}

df = pd.DataFrame(data)
df
```

```python
x = np.arange(len(df['CoursesName']))
bar_width = 0.35

plt.figure(figsize=(9, 6))
plt.grid(True, axis='y', zorder=0)
plt.bar(x - bar_width/2, df['Student Group 1'], width=bar_width, label='Student Group 1', zorder=3, color='deepskyblue')
plt.bar(x + bar_width/2, df['Student Group 2'], width=bar_width, label='Student Group 2', zorder=3, color='beige')
plt.xticks(x, df['CoursesName'], rotation=45)
plt.title('Number of Students per Course by Group')
plt.xlabel('Courses Name')
plt.ylabel('Number of Students')
plt.legend()
plt.show()
```
