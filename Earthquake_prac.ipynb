```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

fig, ax = plt.subplots(figsize=(6,6))

# Outer red iris
outer = plt.Circle((0,0), 1, color='red')
ax.add_patch(outer)

# Inner black pupil
pupil = plt.Circle((0,0), 0.2, color='black')
ax.add_patch(pupil)

# Tomoe positions
angles = np.linspace(0, 2*np.pi, 4)[:-1]
tomoe_patches = []

for angle in angles:
    x = 0.6 * np.cos(angle)
    y = 0.6 * np.sin(angle)
    tomoe = plt.Circle((x,y), 0.1, color='black')
    ax.add_patch(tomoe)
    tomoe_patches.append(tomoe)

ax.set_xlim(-1.2,1.2)
ax.set_ylim(-1.2,1.2)
ax.set_aspect('equal')
ax.axis('off')

# --- ANIMATION FUNCTION ---
def update(frame):
    # Rotate everything around the center
    for patch in ax.patches:
        patch.set_transform(
            ax.transData.rotate_deg_around(0, 0, frame)
        )
    return ax.patches

# --- RUN ANIMATION ---
ani = FuncAnimation(fig, update, frames=360, interval=30)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

fig, ax = plt.subplots(figsize=(6,6))

# --- Base Rinnegan Colors ---
iris_color = "#6A0DAD"   # Purple
pupil_color = "black"
ring_color = "black"

# --- Outer Iris ---
outer = plt.Circle((0,0), 1, color=iris_color)
ax.add_patch(outer)

# --- Pupil ---
pupil = plt.Circle((0,0), 0.15, color=pupil_color)
ax.add_patch(pupil)

# --- Ripple Rings ---
rings = []
ring_radii = [0.30, 0.45, 0.60, 0.75, 0.90]

for r in ring_radii:
    ring = plt.Circle((0,0), r, fill=False, linewidth=3, color=ring_color)
    ax.add_patch(ring)
    rings.append(ring)

# --- Axes Setup ---
ax.set_xlim(-1.2, 1.2)
ax.set_ylim(-1.2, 1.2)
ax.set_aspect('equal')
ax.axis('off')

# --- ANIMATION FUNCTION ---
def update(frame):
    # Rotate the rings slightly for a "moving ripple" effect
    for ring in rings:
        ring.set_transform(
            ax.transData.rotate_deg_around(0, 0, frame * 0.5)
        )
    return rings

# --- RUN ANIMATION ---
ani = FuncAnimation(fig, update, frames=360, interval=30)

plt.show()
```
