```python
# importing the data
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Section 1: Part 1-10
#1).
df = pd.read_csv('Data\woocommerce-product-export.csv') # reading the csv file
```

```python
#2).
df.info() # showing the data frame info
```

```python
#3).
df.describe() # showing the data frame description
```

```python
#4).
df.head() # showing the first 5 rows of the data frame
```

```python
#5).
df.tail() # showing the last 5 rows of the data frame
```

```python
#6).
print(df[['total_profit', 'month_number']]) # showing the total profit and month number columns   
```

```python
#7).

plt.figure(figsize=(10, 5)) # setting the figure size
plt.grid()
plt.bar(df['month_number'], df['total_profit'], color='xkcd:watermelon', zorder=3) # plotting the bar chart
plt.title('Company Profit per Month') # setting the title
plt.xlabel('Month') # setting the x-axis label
plt.ylabel('Total Profit') # setting the y-axis label
plt.show() # displaying the plot
```

```python
#8).
plt.plot(
    df['month_number'],
    df['total_profit'],
    marker='o',
    color='red',
    linestyle=':',
    markerfacecolor='black',
    linewidth=3,
    label='Total Profit',
    zorder=3
)


plt.grid()
plt.xlabel('Month Number') # setting the x-axis label
plt.xticks(df['month_number']) # setting the x-ticks to month numbers
plt.ylabel('Total Profit') # setting the y-axis label
plt.title('Company Profit per Month') # setting the title
plt.legend(loc='lower right', labels=['Profit Data of last year']) # displaying the legend
plt.show() # displaying the plot
```

```python
#9).
products = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']

plt.figure(figsize=(9, 5)) # setting the figure size
plt.grid(True)

for product in products: # using a loop to plot each product's sales data
    plt.plot(
        df['month_number'],
        df[product],
        marker='o',
        linewidth=2,
        label=product.capitalize()
    )
plt.xlabel('Month Number') # setting the x-axis label
plt.xticks(df['month_number']) # setting the x-ticks to month numbers
plt.ylabel('Units Sold') # setting the y-axis label
plt.title('Units Sold per Month') # setting the title
plt.legend(loc='upper left') # displaying the legend
plt.show() # displaying the plot
```

```python
#10). 
plt.figure(figsize=(9, 5)) # setting the figure size
plt.grid(True, linestyle='--') # adding grid lines
plt.scatter(
    df['month_number'],
    df['bathingsoap'],
    color='blue',
    marker='o',
    zorder=3,
)


plt.xlabel('Month Number') # setting the x-axis label
plt.xticks(df['month_number']) # setting the x-ticks to be the month numbers
plt.ylabel('Units Sold') # setting the y-axis label
plt.title('Bathing Soap Sales per Month') # setting the title
plt.legend(['Bathing Soap'], loc='upper left') # displaying the legend
plt.show() # displaying the plot
```

```python
#1).
date=["25/12","26/12","27/12"]
temp=[8.5,10.5,6.8]

plt.figure(figsize=(10, 5)) # setting the figure size
plt.grid() # adding grid lines
plt.plot(date, temp, marker='o', linewidth=2, label='Temperature') # plotting the line chart
plt.xlabel('Date') # setting the x-axis label
plt.ylabel('Temperature') # setting the y-axis label
plt.title('Date-wise Temperature') # setting the title
plt.legend(loc='upper left') # displaying the legend
plt.show() # displaying the plot
```

```python
#2). 
height=[121.9,124.5,129.5,134.6,139.7,147.3,152.4,157.5,162.6]
weight=[19.7,21.3,23.5,25.9,28.5,32.1,35.7,39.6,43.2]

plt.figure(figsize=(9, 5)) 
plt.plot(
    weight,
    height,
    linestyle='-.',
    color='green',
    marker='o',
    markersize=10,
    markerfacecolor='green',
    linewidth=3,
    label='Height vs Weight',
    zorder=3
) # plotting the line chart

plt.grid() # adding grid lines
plt.xlabel('Weight (kg)') # setting the x-axis label
plt.ylabel('Height (cm)') # setting the y-axis label
plt.title('Average weight with respect to Average Height') # setting the title
plt.show()
```
