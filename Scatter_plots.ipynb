```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)
plt.grid()
plt.xlabel('Hours Studied')
plt.ylabel('Scores')
plt.title('Study Hours vs. Scores')
plt.show()
```

```python
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
plt.scatter(x, y, color='red', s=sizes, alpha=0.4)
plt.grid()
plt.show()
```

```python
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))


plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')
plt.grid()
plt.colorbar()
plt.show()
```

```python
AgeGroupOne = range(40, 70, 1)                        # X-values ranging from 40 to 69 year old
Purchase1=np.abs(np.random.randn(30) * 20)

plt.scatter(AgeGroupOne, Purchase1, color='blue', s=100, alpha=0.5, label='Age Group 40-70')


AgeGroupTwo=range(20, 40, 1)    # X-values ranging from 20 to 39 years old
Purchase2=np.abs(np.random.randn(20) * 40)

plt.scatter(AgeGroupTwo, Purchase2, color='orange', s=100, alpha=0.5, label='Age Group 20-40')


plt.legend()
plt.xlabel('Age')
plt.ylabel('Purchase Amount')
plt.title('Purchase Amount by Age Group')
plt.grid(linestyle='--', alpha=0.7)
plt.gca().set_facecolor('lightgray')
plt.show()
```

```python
rng = np.random.RandomState(0) # Creates a random number generator (RNG) object with a specific seed value of 0. This ensures reproducibility of the random numbers generated.
x = rng.randn(100) # Generates an array x of 100 random numbers drawn from a standard normal distribution (mean=0, variance=1).
y = rng.randn(100) # Generates an array y of 100 random numbers drawn from a standard normal distribution (mean=0, variance=1).

colors = rng.rand(100) 
sizes = 1000 * rng.rand(100)


plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral') 
plt.colorbar() 
plt.title('Scatter Plot')
plt.xlabel('Random Data X')
plt.ylabel('Random Data Y')
plt.gca().set_facecolor('lightgray')
plt.grid(linestyle='--', alpha=0.5)  
plt.show()
```
